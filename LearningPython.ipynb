When a variable is declare eg  
***a = 3***

Python will perform three distinct steps to carry out the request.  
These steps reflect the operation of all assignments in the Python language:  
1. Create an object to represent the value 3.  
2. Create the variable a, if it does not yet exist.  
3. Link the variable a to the new object 3.  

The variable name and the object are linked using pointers aka **references**  

a-->objext(3)

Internally, the variable is really a pointer to the object’s memory space created by running the literal expression 3.

In simple terms;  
• Variables are entries in a system table, with spaces for links to objects.  
• Objects are pieces of allocated memory, with enough space to represent the values for which they stand.  
• References are automatically followed pointers from variables to objects.  

Each object also has two standard header fields: a ***type designator*** used to mark the ***type of the object***, and a ***reference counter*** used to determine when it’s OK to reclaim the object.  

Types live within objects not variables. Objects have no types.
a = 3 # It's an integer
a = 'spam' # Now it's a string
a = 1.23
In this case were simply changing the reference of the variable (a) to a different object ie of type int to string as recorden in the object's type designator.

When a variable is declared python internally uses a counter to how many times an object is referenced and when that counter drops to zero, python's garbage collector will clean up the memory space unless it's cached.

### Lists


Lists are Python’s most flexible ordered collection object type. Unlike strings, lists can contain any sort of object: numbers, strings, and even other lists. Also, unlike strings, lists may be **changed in-place** by assignment to offsets and slices, list method calls, deletion statements, and more—they are mutable objects.  

The main list manipulations are the same as those of strings eg concatenation etc except they return new List objects.  

**MAP** --> apply a function to an iterable object multiple times.  

    map(ord,myList)

One warning here: beware that append and sort change the associated list object inplace, but *don’t return the list as a result* (technically, they both return a value called None). If you say something like L=L.append(X), you won’t get the modified value of L (in fact, you’ll lose the reference to the list altogether!). When you use attributes such
as append and sort, objects are changed as a side effect, so there’s no reason to reassign.  

Like strings, lists have other methods that perform other specialized operations. For instance, reverse reverses the list in-place, and the extend and pop methods insert multiple items at the end of and delete an item from the end of the list, respectively.

### Tuples


Tuples construct simple groups of objects. They work exactly like lists, except that tuples can’t be changed inplace (they’re immutable) and are usually written as a series of items in parentheses.
The can support arbitrary types and nesting, are sequences, immutable, accessed by offset and are relatively quick to index.  
They can support many of the list ops eg concatenation, repetition, indexing and slicing etc but cannot be altered in place.However the contents eg a list in tuple can be changed

### Scope and nested functions

Enclosing scopes are also know as statically nested scopes.

Global declarations make the search begin in the global (module file) scope instead.  

Nesting function can achieve a **closure or factory function.** These terms refer to a function object that remembers values in enclosing scopes regardless of whether those scopes are still present in memory as shown in the **maker fn** below. 

>The most unusual part of this is that the nested function remembers integer 2, the value of the variable N in maker,  
>even though maker has returned and exited by the time we call action.   
>In effect,N from the enclosing local scope is retained as state information attached to action, and we get back  its argument squared.    

Calling maker again(g) retuns a new object with different state info than f aka *g remembers 3, and f remembers 2.*

````python
def maker(N):
    def action(X):         # Make and return action
        return X ** N      # action retains N from enclosing scope
    return action


f=maker(2)
print(f)   #return a fn object of the action
print(f(3))  #remembers the 2 and then raises the 3 to it and we get 9

g = maker(3) # g remembers 3, f remembers 2
print(g(3)) # 3 ** 3 =27

print(f(3))  #print 9 different from the 27
#both f and g have different state information
````

Generator expressions are similar to the list comprehensions of the prior section, but they return an object that produces results on demand instead of building a  result list.

Send is like next but allows communication with the caller.

# 🔹 1. What is Composition?

- **Inheritance** is an “_is-a_” relationship.
  Example: A `Dog` **is an** `Animal`.
- **Composition** is a “_has-a_” relationship.
  Example: A `Car` **has an** `Engine`.

Instead of inheriting from another class, one class uses another class **as a component** (attribute).
This allows building **complex objects** by combining simpler ones.

---

# 🔹 2. Composite Objects

A **composite object** is simply an object that contains other objects.
It’s a whole made up of parts.
For example, a `Car` object may be composed of `Engine`, `Wheel`, and `Seat` objects.

# 🔹 3. Example: Composition in Python

```python

        print(f"Engine with {self.horsepower} HP starting...")

class Wheel:
    def __init__(self, size):
        self.size = size

    def rotate(self):
        print(f"Wheel of size {self.size} is rotating...")

class Car:
    def __init__(self, brand, horsepower, wheel_size):

        # composition: car has an engine and wheels
        self.engine = Engine(horsepower)
        self.wheels = [Wheel(wheel_size) for _ in range(4)]

        self.engine.start()
        for wheel in self.wheels:
            wheel.rotate()
        print(f"{self.brand} is driving!")
```

✅ **Flexibility** – You can swap components without rewriting the whole class (e.g., change `Engine` to `ElectricEngine`).
✅ **Code reuse** – Use small classes in different contexts.
✅ **Avoid deep inheritance chains** – Composition is often preferred over inheritance (“**favor composition over inheritance**”).

Both involve “_has-a_”:

- **Composition**: Strong ownership. Parts cannot exist without the whole. (E.g., a `Car`’s `Engine` dies with the car.)
- **Aggregation**: Weaker relationship. Parts can exist independently. (E.g., a `Team` has `Players`, but players can exist without the team.)

The `__getattr__` method in Python is a special method that allows you to define a fallback mechanism for accessing attributes. It's automatically called when you try to access a nonexistent attribute on an object.

-----

## How it works

When you try to access an attribute on an object, Python first checks the object's dictionary (`__dict__`) for that attribute. If the attribute isn't found, Python then checks the class's method resolution order (MRO) for the attribute. If it's still not found, Python looks for the `__getattr__` method in the object's class. If found, it calls `__getattr__` with the name of the missing attribute as an argument.

The `__getattr__` method takes one argument: the name of the attribute you're trying to access as a **string**. You can then define the logic to handle the missing attribute, like returning a default value, raising a different exception, or dynamically creating the attribute.

-----

## Example

Here's a simple example of how `__getattr__` can be used.

```python
class MyClass:
    def __init__(self):
        self.existing_attribute = "Hello"

    def __getattr__(self, name):
        print(f"__getattr__ called for attribute '{name}'")
        return f"This attribute '{name}' doesn't exist."

# Create an instance of MyClass
obj = MyClass()

# Access an existing attribute
print(f"Existing attribute: {obj.existing_attribute}")
# Output: Existing attribute: Hello

# Access a nonexistent attribute
print(f"Nonexistent attribute: {obj.nonexistent_attribute}")
# Output: __getattr__ called for attribute 'nonexistent_attribute'
# Nonexistent attribute: This attribute 'nonexistent_attribute' doesn't exist.
```

In this example, when we try to access `obj.nonexistent_attribute`, Python can't find it. So it calls `__getattr__`, which prints a message and returns a string, effectively handling the missing attribute gracefully.

-----

## Key difference between `__getattr__` and `__getattribute__`

It's important to distinguish between `__getattr__` and `__getattribute__`.

  * `__getattr__` is only called for **nonexistent** attributes.
  * `__getattribute__` is called for **all** attribute accesses, whether they exist or not. Using `__getattribute__` requires more caution because it can easily lead to infinite recursion if not implemented carefully.

For most use cases, `__getattr__` is the preferred choice as it's a safer and more specific way to handle missing attributes.

```python
            return value
        else:
```

# 1. `__getattr__` vs `__getattribute__`

Both are **special methods** for customizing **attribute access** in Python objects.

* **`__getattr__(self, name)`**
  Called **only when the attribute is NOT found normally** (fallback).

* **`__getattribute__(self, name)`**
  Called **every single time an attribute is accessed**, whether it exists or not.
  ⚠️ Must be careful, or you’ll cause **infinite recursion**.

### Example: `__getattr__`

```python
class MyClass:
    def __init__(self):
        self.x = 10

    def __getattr__(self, name):
        # Only called when attribute not found
        print(f"__getattr__ triggered for: {name}")
        return f"{name} not found!"

obj = MyClass()
print(obj.x)         # Found normally → prints 10
print(obj.y)         # Not found → triggers __getattr__
```

**Output:**

```
10
__getattr__ triggered for: y
y not found!
```

---

### Example: `__getattribute__`

```python
class MyClass:
    def __init__(self):
        self.x = 42

    def __getattribute__(self, name):
        print(f"__getattribute__ triggered for: {name}")
        # Use super() to avoid infinite recursion
        return super().__getattribute__(name)

obj = MyClass()
print(obj.x)         # Always triggers __getattribute__
```

**Output:**

```
__getattribute__ triggered for: x
42
```

👉 Summary:

* Use `__getattr__` when you want a fallback for missing attributes.
* Use `__getattribute__` when you want *complete control* over attribute access.

# 2. **Properties** (`property()`)

Properties let you **control access to attributes** without changing how they’re used.
This is Python’s way of doing **getters and setters** (like in Java/C++), but more Pythonic.

### Example with `property`

```python
class Person:
    def __init__(self, name):
        self._name = name  # internal attribute

    def get_name(self):
        print("Getting name...")
        return self._name

    def set_name(self, value):
        print("Setting name...")
        if not value:
            raise ValueError("Name cannot be empty!")
        self._name = value

    name = property(get_name, set_name)

p = Person("Alice")
print(p.name)       # Calls get_name
p.name = "Bob"      # Calls set_name
print(p.name)
```

**Output:**

```
Getting name...
Alice
Setting name...
Getting name...
Bob
```

👉 Properties allow you to make **attributes act like methods**, but accessed like normal fields.

# 3. **Descriptors**

Descriptors are a **lower-level mechanism** that powers properties, methods, and more.
A descriptor is any class that defines one or more of these methods:

* `__get__(self, instance, owner)` → controls reading
* `__set__(self, instance, value)` → controls writing
* `__delete__(self, instance)` → controls deleting

### Example: Custom Descriptor

```python
class MyDescriptor:
    def __get__(self, instance, owner):
        print("Getting value...")
        return instance._value

    def __set__(self, instance, value):
        print("Setting value...")
        if value < 0:
            raise ValueError("Value must be >= 0")
        instance._value = value

class MyClass:
    attr = MyDescriptor()   # attr is managed by descriptor

    def __init__(self, value):
        self._value = value

obj = MyClass(10)
print(obj.attr)     # Calls __get__
obj.attr = 20       # Calls __set__
print(obj.attr)
```

**Output:**

```
Getting value...
10
Setting value...
Getting value...
20
```

👉 Notes:

* `property()` is just a **convenience wrapper around a descriptor**.
* Descriptors let you build **reusable logic for attributes** (validation, logging, lazy loading, etc.).

* **`__getattr__`** → fallback when attribute doesn’t exist.
* **`__getattribute__`** → intercepts *all* attribute access.
* **Properties** → high-level way to make getters/setters look like attributes.
* **Descriptors** → low-level protocol to manage attributes (properties are just a built-in descriptor).

# 🔹 What are Metaclasses?

* In Python, **everything is an object**, even **classes themselves**.
* Normal objects are created from **classes**.

* But **who creates classes?** → That’s the job of a **metaclass**.

👉 A **metaclass is the “class of a class.”**
Just like a class defines how instances are created, a metaclass defines how classes are created.

---

# 🔹 Default Metaclass in Python

By default, every class in Python is created by the built-in metaclass called **`type`**.

```python

    pass

print(type(MyClass))   # <class 'type'>
print(isinstance(MyClass, type))  # True
```

---

# 🔹 Why use Metaclasses?

Metaclasses allow you to:

1. **Control class creation** (e.g., modify attributes, enforce coding rules).
2. **Automatically add methods/properties to classes.**
3. **Validate class definitions.**

Think of them as "hooks" into class creation.

---

# 🔹 How to Create a Metaclass

A metaclass is just a **class that inherits from `type`**.

```python
# Define a simple metaclass
class MyMeta(type):
    def __new__(cls, name, bases, dct):
        print(f"Creating class {name} with metaclass {cls.__name__}")
        return super().__new__(cls, name, bases, dct)

# Use the metaclass in a class
class MyClass(metaclass=MyMeta):
    pass
```

📌 Output:

```
Creating class MyClass with metaclass MyMeta
```

👉 Here, `__new__` of the metaclass runs when the class itself is created.

# 🔹 Adding Rules with Metaclasses

You can use metaclasses to **enforce rules** on class definitions.

```python
class NoLowerCaseAttributes(type):
    def __new__(cls, name, bases, dct):
        for attr in dct:
            if attr.islower():
                raise ValueError(f"Attribute '{attr}' in {name} must be UPPERCASE")
        return super().__new__(cls, name, bases, dct)

# This will raise an error
try:
    class BadClass(metaclass=NoLowerCaseAttributes):
        myattr = 10   # ❌ lowercase
except Exception as e:

```

# 🔹 Automatically Adding Methods

Metaclasses can also inject methods automatically.

```python
                return f"<{name} object with attrs={self.__dict__}>"
```
